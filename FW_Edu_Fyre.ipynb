```python
root_path = '/content/drive/MyDrive/FW/Models/'
model_path = f'{root_path}/model'
tokenizer_path = f'{root_path}/tokenizer'
model_name = 'facebook/nllb-200-distilled-600M'
```

```python
def load_model_and_tokenizer(model_name, model_path, tokenizer_path):
  print('Loading model...')

  if os.path.exists(model_path) or os.path.exists(tokenizer_path):
    print('Loading model from local storage.')
    model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_path)
  else:
    print('Downloading NLLB model from huggingface model garden.')
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    model.save_pretrained(model_path)
    tokenizer.save_pretrained(tokenizer_path)

  print('Model and tokenizer successfully loaded.')
  return model, tokenizer
```

```python
load_model_and_tokenizer(model_name, model_path, tokenizer_path)
```

```python
model, tokenizer = load_model_and_tokenizer(model_name, model_path, tokenizer_path)
```

```python
def translate(text, src_language, tgt_language):
  translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang=src_language, tgt_lang=tgt_language)
  return translator(text, max_length=512)
```

```python
import time

texts = [
    """How is bipolar disorder different from unipolar depression or 'regular' depression? Answer: Both bipolar disorder and major depression are typically associated with depressive episodes. So both illnesses are accompanied by depressions. The difference is that in bipolar disorder people also have periods of elevation -- or severe irritability. We call these manic or hypomanic episodes."""
]

translation = '';

start = time.time()
for text in texts:
  translation = translate(text, "eng_Latn", "sna_Latn")[0]['translation_text']
end = time.time()

print(f'estimated T == {end-start}\n\n{translation}')
```
